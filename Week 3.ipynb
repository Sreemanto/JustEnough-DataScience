```python
#lets declare the size of an array
size = 10000000

#Creating two generators for lists
list1 = range(size)
list2 = range(size)

#Create two numpy arrays
arr1 = np.arange(size)
arr2 = np.arange(size)

#calculate time for adding these two array into a list
start = time.time()
result1 = [(x+y) for x,y in zip(list1,list2)]
print("Time for python List in msec : ",(time.time() - start))


#calculating time for numpy array
start = time.time()
result2 = (arr1 + arr2)
print("Time for Numpy array in msec : ",(time.time() - start))
```

#### q1. How to extract all numbers between a given range from a numpy array

```python
a = np.arange(50)
```

```python
#Method 1 (15-30)
out_arr = a[(a>=15) & (a<=30)]
out_arr
```

```python
#Method 2
index = np.where((a>=15) & (a<=30))
#index
out_arr = a[index]
out_arr
```

#### Q3. How to replace all values greater than a given value to a certain cutoff

```python
a = np.random.randint(0,80,100)
```

```python
#1st method - use clip 
out_array = np.clip(a,a_min=18,a_max=60)
out_array
```

```python
#2nd Method - use where
np.where(a<18, 18, np.where(a>60,60,a))
```
